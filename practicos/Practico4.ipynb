```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lumfunc(m,p):
    tmp = -0.4*(m-p[1])
    f = 0.4*np.log(10)*p[0]*np.power(10.0,tmp*(p[2]+1))
    f = f*np.exp(-np.power(10.0,tmp))
    return(f)
```

```python
filename = "../material/LF_Blanton2001.csv"
l,f = np.genfromtxt(filename,unpack=True,skip_header=1,delimiter=',')
plt.plot(l,f,'*')

x = np.linspace(l.min(),l.max(),100)
y = lumfunc(x,[1.46e-2,-20.83,-1.20])
plt.plot(x,y)

plt.semilogy()
```
